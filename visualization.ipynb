Зачастую признак может быть зашумлен или может содержать лишнюю (слишком детализированную, ведущую к переобучению) информацию. Возможным решением этой проблемы служит бинаризация признака.

Бинаризуем признаки Age и Fare

* pd.cut - разбиение целочисленных данных на несколько интервалов по квантилям
* pd.qcut - разбиение числовых (не обязательно целочисленных) данных на несколько интервалов по квантилям

```python
Data['AgeBin'] = pd.cut(Data['Age'].astype(int), 5)
Data['AgeBin'].head()
```

Теперь переведём полученные интервалы в числа, используя LabelEncoder. Полученные метки уже можно использовать как числовые признаки.

```python
from sklearn.preprocessing import LabelEncoder
label = LabelEncoder()
Data['AgeBin_Code'] = label.fit_transform(Data['AgeBin'])
Data[['Age','AgeBin','AgeBin_Code']].head()
```

Из графика видно, что следующие группы мужчин выжили (точно): 

1) Пассажиры, чей FareBin_Code = 1, из возрастных групп 0 и 6.

2) Пассажиры, чей FareBin_Code = 3, из возрастной группы 4.

Соберём все полученные выше условия и посмотрим на улучшение.

```python
def second_prediction(x):
    # обрабатываем женщин
    if x.Sex == 0:
        if x.Pclass == 1 and x.AgeBin_Code2 == 0:
            return 0
        if x.Pclass == 3 and x.AgeBin_Code2 >= 5 and x.AgeBin_Code2 <= 6:
            return 0
        if x.Pclass == 2 and x.AgeBin_Code2 > 6:
            return 0
        if x.FamilySize > 4 and x.AgeBin_Code2 == 3:
            return 0
        return 1
    # обрабатываем мужчин
    if (x.Pclass != 3 or x.FamilySize >= 2 and x.FamilySize <= 4) and x.AgeBin_Code2 == 0:
        return 1
    if (x.Pclass == 1 or x.FamilySize == 2 or x.FamilySize == 4 ) and x.AgeBin_Code2 == 1:
        return 1
    if x.AgeBin_Code2 == 4 and x.FareBin_Code == 3:
        return 1
    if (x.AgeBin_Code2 == 6 or x.AgeBin_Code2 == 0) and x.FareBin_Code == 1:
        return 1
    if x.Pclass == 1 and x.AgeBin_Code2 == 9:
        return 1
    return 0
```

```python
Data['prediction2'] = Data.apply(lambda x: second_prediction(x), axis=1)
```

```python
accuracy_score(Data['Survived'], Data['prediction2'])
```

```python
accuracy_score(Data['Survived'], Data['prediction2']) - accuracy_score(Data['Survived'], Data['prediction'])
```
